```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('../data/SIMD_2020_Data.csv')
print(df)
```

```python
df['id']=df['id'].fillna(value=0)
df['length_more_than_5'] = df['id'].apply(lambda x: len(str(x)) > 5)
df
```

```python
indices = df[df['length_more_than_5']].index

# Shift data one column to the right for rows where 'length_more_than_5' is True
for index in indices:
    df.iloc[index, 1:] = df.iloc[index, :-1]


```

```python
missing_data_rows = df[df['id'].str.len() > 5]
print(missing_data_rows)
```

```python
cleaned_df=df[df['id'].str.len() <= 5]
print(cleaned_df)
```

```python
excel_file_path = '../data/cleaned_2020.xlsx'
# Writing the DataFrame to the Excel file
cleaned_df.to_excel(excel_file_path, index=False)
```
