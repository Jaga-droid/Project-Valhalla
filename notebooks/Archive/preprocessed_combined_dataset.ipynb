```python
import pandas as pd
```

SIMD 2016 Data

```python
df_2016=pd.read_excel('../data/SIMD_2016_Data.xlsx')
```

SIMD 2020 Data

```python
df_2020=pd.read_excel('../data/cleaned_2020.xlsx')
```

Dropping unnecessary columns from SIMD 2020 dataframe

```python
df_2020=df_2020.drop(columns=['id','geom','simd2020v2','simd_2020v','simd2020_1','simd2020_2','simd2020_3','simd2020_4','simd2020_e','simd2020_h','simd2020_5','simd2020_a','simd2020_c','simd2020_6','broadband','length_more_than_5'])
```

Creating Data Dictionary for mapping header names

```python
header_map = {

    'datazone':'Data_Zone',
    'intermedia':'Intermediate_Zone',
    'council_ar':'Council_area',
    'total_popu':'Total_population',
    'working_ag':'Working_age_population_revised',
    'income_rat':'Income_rate',
    'income_cou':'Income_count',
    'employment':'Employment_rate',
    'employme_1':'Employment_count',
    'cif':'CIF',
    'alcohol':'ALCOHOL',
    'drug':'DRUG',
    'smr':'SMR',
    'depress':'DEPRESS',
    'lbwt':'LBWT',
    'emerg':'EMERG',
    'attendance':'Attendance',
    'attainment':'Attainment',
    'no_qualifi':'Noquals',
    'not_partic':'NEET',
    'university':'HESA',
    'drive_petr':'drive_petrol',
    'drive_gp':'drive_GP',
    'drive_post':'drive_PO',
    'drive_prim':'drive_primary',
    'drive_reta':'drive_retail',
    'drive_seco':'drive_secondary',
    'pt_gp':'PT_GP',
    'pt_post':'PT_Post',
    'pt_retail':'PT_retail',
    'crime_coun':'crime_count',
    'crime_rate':'crime_rate',
    'overcrowde':'overcrowded_count',
    'nocentralh':'nocentralheat_count',
    'overcrow_1':'overcrowded_rate',
    'nocentra_1':'nocentralheat_rate'
  }
```

Renaming headers in SIMD 2020 dataframe

```python
df_2020 = df_2020.rename(columns=header_map)
```

```python
df_2020
```

Adding column "year" in both 2016 and 2020 dataframe

```python
df_2016['year'] = 2016
df_2016['year'] = df_2016['year'].astype('int')
```

```python
df_2020['year'] = 2020
df_2020['year'] = df_2020['year'].astype('int')
```

Combining both dataframes

```python
combined_df = pd.concat([df_2016, df_2020], ignore_index=True)
```

```python
combined_df
```

Exporting the combined dataframe to csv file

```python
combined_df.to_csv('../data/combined_dataset.csv', index=False, header=True)
```
